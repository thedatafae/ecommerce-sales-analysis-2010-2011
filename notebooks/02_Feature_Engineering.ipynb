# 2. Feature Engineering

```python
import numpy as np 
import pandas as pd 
```

```python
data = pd.read_csv("cleaned_data.csv")
df = pd.DataFrame(data)
df.head()
```

```python
# Net sales per row
df['net_sales'] = df['quantity'] * df['unitprice']
```

```python
# Flag returns
df['is_return'] = df['quantity'] < 0
```

```python
df['invoicedate'] = pd.to_datetime(df['invoicedate'])
```

```python
# Extract month/year from invoice_date 
df['invoice_month'] = df['invoicedate'].dt.to_period('M')
df['invoice_year'] = df['invoicedate'].dt.to_period('Y')
```

```python
# weekday or hour for purchase patterns
df['invoice_day'] = df['invoicedate'].dt.day_name()
df['invoice_hour'] = df['invoicedate'].dt.hour
```

```python
df.to_csv('featured_data.csv', index=False)
```
