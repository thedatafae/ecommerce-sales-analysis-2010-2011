# 1. Data Cleaning

```python
import numpy as np 
import pandas as pd 
```

```python
data = pd.read_csv("data.csv")
df = pd.DataFrame(data)
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.columns = df.columns.str.lower()
```

```python
df['is_cancelled'] = df['invoiceno'].str.startswith("C")
```

```python
df['invoicedate'] = pd.to_datetime(df['invoicedate'])
```

```python
df['description'] = df['description'].fillna('Unknown')
```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates()
```

```python
bad_desc = ['Unknown', '?', 'faulty', 'damages', None]

df.loc[:, 'description'] = df.apply(lambda x: x['description'] if x['description'] not in bad_desc else x['stockcode'], axis=1)
```

```python
df.to_csv('cleaned_data.csv', index=False)
```
